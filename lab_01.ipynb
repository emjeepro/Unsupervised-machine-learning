```python
# подключим необходимые библиотеки
import warnings
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_samples
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score, adjusted_rand_score
warnings.filterwarnings('ignore')
```

```python
def optimal_clusters(X, cluster, kmeans):
    fig, (ax1) = plt.subplots(1, 1)
    fig.set_size_inches(9, 9)
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (cluster + 1) * 10])
    y_lower = 10
    sample_silhouette = silhouette_samples(X, kmeans.labels_)  # вычислить силуэт для каждой точки
    for i in range(cluster):
        i_silhouette = sample_silhouette[kmeans.labels_ == i]
        i_silhouette.sort()
        size_i = i_silhouette.shape[0]
        y_upper = y_lower + size_i
        color = cm.nipy_spectral(float(i) / cluster)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, i_silhouette, facecolor=color, edgecolor=color, alpha=0.7)
        y_lower = y_upper + 10

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")
    ax1.axvline(x=silhouette_score(X, y_pred), color="red", linestyle="--")  # the vertical line for average silhouette score of all the values
    ax1.set_yticks([])  # clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.show()
```

```python
from sklearn.metrics import silhouette_score

variances = [1.0, 2.5, 0.5]

# iterate over the variances
for variance in variances:
    # 1-й этап: сгенерируйте датасет в виде облака из 1500 точек с тремя центрами и дисперсиями 1.0, 2.5, 0.5 (параметр cluster_std)
    X, y = make_blobs(n_samples=1500, centers=3, cluster_std=variance)

    # plot the dataset
    plt.scatter(X[:, 0], X[:, 1], c="black")
    plt.title(f'original_{variance}.png')
    plt.show()

    for i in range(2, 6):
        # 2-й этап: обучите алгоритм кластеризации K-means (класс sklearn.cluster.KMeans)
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(X)

        # predict the quantity of clusters
        y_pred = kmeans.predict(X)

        # plot the data with clusters
        plt.scatter(X[:, 0], X[:, 1], c=y_pred)
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=100)
        plt.title(f'KMEANS_{variance}_{i}.png')
        plt.show()

        # 3-й этап: вычислите для каждой модели метрики: однородность, полнота, V-мера, ARI, силуэт
        print('Homogeneity: ', homogeneity_score(y, y_pred))
        print('Completeness: ', completeness_score(y, y_pred))
        print('V-measure: ', v_measure_score(y, y_pred))
        print('Adjusted Rand Index: ', adjusted_rand_score(y, y_pred))
        print('Silhouette Coefficient: ', silhouette_score(X, y_pred))
        print()

        # 4-й этап: для оценки оптимального числа кластеров воспользуйтесь графиком зависимости количество точек от значений коэффициента силуэта
        optimal_clusters(X, i, kmeans)

        # 5-й этап: вычислите для каждой модели параметр kmeans.inertia_ и постройте график его зависимости от числа кластеров в интервале 1-8
        inertia = []
        for j in range(1, 9):
            kmeans = KMeans(n_clusters=j, random_state=42)
            kmeans.fit(X)
            inertia.append(kmeans.inertia_)
        plt.plot(range(1, 9), inertia)
        plt.title(f'Inertia plot')
        plt.show()
```
